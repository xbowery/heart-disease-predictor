<h3>III. Methodology</h3>

<br>
<p>
<b>Concatenate Datasets:</b><br>
We merged the datasets (via concatenation) of the four processed data from Cleveland, Switzerland, Hungary and Long Beach. This gave a total of 920 data entries, with the same 14 attributes above. <br><br>

<b>Missing Column Headers:</b><br>
As the datasets did not have columns headers, we added column headers to the 14 attributes in the combined data set.<br><br>

<b>Data Cleaning - Achieving Data Consistency:</b><br>
During the preliminary load, the data consisted of a mixture of string, integer and float datatypes. <br>
Datatype for all attributes were amended to float in our dataset.<br>
All missing values with "?" were replaced as nan to allow us to count missing values. </p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data set
df1 = pd.read_csv ('processed.cleveland.data',header=None)
df2 = pd.read_csv ('processed.switzerland.data',header=None)
df3 = pd.read_csv ('processed.hungarian.data',header=None)
df4 = pd.read_csv ('processed.va.data',header=None)

df = pd.concat([df1,df2,df3,df4])
df.columns=['age','sex','chest_pain_type','resting_blood_pressure','cholesterol','fasting_blood_sugar',
            'resting_electrocardiographic_results','maximum_heart_rate','exercise_induced_angina',
            'ST_depression',"slope_peak_exercise",'no_major_vessels_colored','thal','diagnosis_of_heart_disease']

# replace all the values with "?" as nan for counting of missing values.
df = df.replace("?", np.nan)

#convert all the data in the dataframe into float.
df = df.astype(float)
```

<b>Data Cleaning - Observing Missing Data:</b><br>
We can now clearly observe the distribution of missing data across each attribute.


```python
#total percentage of missing data
missing_data = df.isnull().sum()
total_percentage = (missing_data.sum()/(df.shape[0]*len(df.columns))) * 100
print(f'The total percentage of missing data is {round(total_percentage,2)}%')
# percentage of missing data per category
total = df.isnull().sum().sort_values(ascending=False)
percent_total = round((df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)*100 ,2)
missing = pd.concat([total, percent_total], axis=1, keys=["Total", "Percentage"])
missing_data = missing[missing['Total']>0]

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(24,6))
sns.set(style="whitegrid")
sns.barplot(x=missing_data.index, y=missing_data['Percentage'], data = missing_data)
plt.title('Percentage of missing data by feature')
plt.xlabel('Features', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.show()

print(missing_data)
```

<b>Data Cleaning - Dropping attributes with High % of Missing Data:</b><br>
The attributes with the most missing data were ``no_major_vessels_colored`` (66.41%) and  `thal` (52.83%). With more than 50% data missing, we decided to drop these 2 attributes in our analysis.<br><br>
This leaves our dataset with 12 attributes, where 11 will be tested to predict `diagnosis_of_heart_disease`.

```python
df.drop(['no_major_vessels_colored', 'thal'], axis='columns', inplace=True)
```

<b>Data Cleaning - Replacing Missing Values:</b><br>
<p>
The remaining missing values were found in both <b>continuous</b> and <b>categorical</b> data that were dealt with by value replacement.<br><br>

For continuous data, missing values were replaced with its <b>mean value</b> for the following:<br>
- `resting_blood_pressure`<br>
- `cholesterol`<br>
- `maximum_heart_rate`<br>
- `ST_depression`<br>

For categorical data, missing values were replaced with <b>"0"</b> for the following:<br>
- `fasting_blood_sugar`<br>
- `exercise_induced_angina`<br>
- `resting_electrocardiographic_results` (to be overridden by One-Hot Encoding below)<br>
- `slope_peak_exercise` (to be overridden by One-Hot Encoding below)<br>
<br>


<b>Data Cleaning - Changing Outcome to Binary Form:</b><br>
For the final outcome on `diagnosis_of_heart_disease`, results were a range of "0-4" where "0" indicated no heart disease, and "1-4" indicated different severities of heart disease. Thus, values "1-4" were replaced with <b>"1"</b> as each of these results were representative of having heart disease, thereby making the outcome in binary form.

</p>

```python
df['resting_blood_pressure'].replace(0, np.nan, inplace=True)
df['resting_blood_pressure'].fillna(int(df['resting_blood_pressure'].mean()), inplace=True)

df['cholesterol'].replace(0, np.nan, inplace=True)
df['cholesterol'].fillna(int(df['cholesterol'].mean()), inplace=True)

df['maximum_heart_rate'].fillna(int(df['maximum_heart_rate'].mean()), inplace=True)
df['ST_depression'].fillna(int(df['ST_depression'].mean()), inplace=True)

df['fasting_blood_sugar'].fillna(value=0, inplace=True)
df['resting_electrocardiographic_results'].fillna(value=0, inplace=True)
df['exercise_induced_angina'].fillna(value=0, inplace=True)
df['slope_peak_exercise'].fillna(value=0, inplace=True)

replace_table = { 1 : 1, 2 : 1, 3 : 1, 4 : 1}

# to convert the diagnosis (outcome) to 2 states instead of multiple states / levels
# State Meaning :  0 is 'Healthy', {1 ,2,3,4} becomes 1 and is 'Sick'}

df.diagnosis_of_heart_disease = df.diagnosis_of_heart_disease.replace(replace_table)  

df[['diagnosis_of_heart_disease']].astype(str).value_counts(normalize=True).plot(kind='bar')
plt.xticks(rotation=0)

plt.show()

```

<b>a) <u>Multivariate Analysis</u></b><br>
A multivariate analysis looks at more than two variables, which aims to find patterns and correlations between several variables simultaneously. This is suitable for our dataset with multiple risk attributes provided to identify heart disease. In contrast to a bivariate analysis, this method would allow for a much deeper and complex understanding of a scenario.<br>
The result of this technique is a correlation matrix that helps us identify multicollinearity (where one predictor variable can be linearly predicted with other predictor variables where coefficients > 0.8 signal multicollinearity).<br>

```python
import seaborn as sns

f, ax = plt.subplots(figsize=(15,15))

df_corr = df.corr()

sns.heatmap(df_corr,
            mask=np.zeros_like(df_corr, dtype=np.bool), 
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax,annot=True);

```

<b>Results:</b><br>
From the correlation matrix produced, there are no coefficients above 0.8, suggesting that the multicolinearity amongst attributes is weak.<br>
Attributes with the highest correlations were below 0.7 such as:
- `slope_peak_exercise` and `ST_depression` (0.61)
- `chest_pain_type` and `diagnosis_of_heart_disease` (0.47) 

<b>b) <u>Decision Tree Confusion Matrix (Full Tree vs. Pruned Tree)</u></b><br>

Decision Trees automatically learn from the dataset to separate records into subgroups by creating splits on predictors. These splits create rules that are transparent and easily understandable. We create both full and pruned trees which have different depths of splitting rules. 
<br>
The result is a confusion matrix that helps us calculate the reliability of the model. Afterwhich, we plot a bar graph of important features that help us identify the best predictors of heart disease. We can also compare which of the two trees are more reliable in predicting heart disease.
<br>

```python
# Create Decision-Tree 

# 11 shortlisted features, excluding last column 'diagnosis_of_heart_disease'

features = [
    'age',
    'sex',
    'chest_pain_type',
    'resting_blood_pressure',
    'cholesterol',
    'fasting_blood_sugar',
    'resting_electrocardiographic_results',
    'maximum_heart_rate',
    'exercise_induced_angina',
    'ST_depression',
    'slope_peak_exercise'
]   

##### define x and y


df_x = df[features]
df_y = df['diagnosis_of_heart_disease']

#display(df_x, df_y)

# train,test,split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df_x, df_y,   
    random_state=0, 
    test_size=0.2
)
    
# x_train.head()
# y_train.head()
y_test.shape

```

```python
# Training Full & Pruned Tree  

# Full Tree #df_clf_fulltree
#train decision tree using DecisionTreeClassifier;

from sklearn.tree import DecisionTreeClassifier

df_clf_fulltree = DecisionTreeClassifier(criterion='entropy', random_state = 0) # or gini
df_clf_fulltree.fit(x_train, y_train)
print("Full Tree Depth :",df_clf_fulltree.tree_.max_depth)

# Pruned Tree #df_clf_prunedtree

from sklearn.tree import DecisionTreeClassifier

df_clf_prunedtree = DecisionTreeClassifier(criterion='entropy',max_depth=4, random_state = 0) # or gini
df_clf_prunedtree.fit(x_train, y_train)
print("Pruned Tree Depth :",df_clf_prunedtree.tree_.max_depth)
```

```python
# Decision Tree - Test Results & Create Confusion Matrix

#test results using .predict()

full_results = df_clf_fulltree.predict(x_test)
# print("Full tree predictions:")
# print(full_results)

# print("-----")

pruned_results = df_clf_prunedtree.predict(x_test)
# print("Pruned tree predictions:")
# print(pruned_results)

# Create confusion matrix

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm_f = confusion_matrix(y_test, full_results) # actual labels, predicted labels
cm_p = confusion_matrix(y_test, pruned_results)

#print(cm_f) 
#print(cm_p)

from sklearn.metrics import plot_confusion_matrix

titles_options = [("Confusion matrix (Full tree)", df_clf_fulltree),
                  ("Confusion matrix (Pruned Tree)", df_clf_prunedtree)]

for title, classifier in titles_options:
    disp = plot_confusion_matrix(classifier, x_test, y_test,
                                 cmap=plt.cm.Blues, values_format='d',
                                 normalize=None)

    disp.ax_.set_title(title)


plt.show()
```

```python
# Decision Tree Model Reliability Full vs Pruned
# Calculate reliability of Model

# Recall no heart disease = TN / (TN + FP) # divide by row
# Recall heart disease = TP / (TP + FN)
# Accuracy = (TP+TN) / (P+N)
# Precision = TP / (TP+FP) # divide by column

###### Full


TNA0P0_F = cm_f[0][0]
FPA0P1_F = cm_f[0][1]
FNA1P0_F = cm_f[1][0]
TPA1P1_F = cm_f[1][1]

# Recall
recall_predict_noheartdisease_full = TNA0P0_F / (TNA0P0_F + FPA0P1_F) # divide by true =0
recall_predict_heartdisease_full = TPA1P1_F /  (TPA1P1_F + FNA1P0_F) # divide by true =1

print('Full Tree Recall Predict No Heart Disease:', round(recall_predict_noheartdisease_full,2)) # predict no heart disease
print('Full Tree Recall Predict Heart Disease:', round(recall_predict_heartdisease_full,2)) #predict heart disease

# Accuracy
print('Full Tree Accuracy:', round(df_clf_fulltree.score(x_test, y_test),2))

# Precision 
print('Full Tree Precision:', round(TPA1P1_F/(TPA1P1_F + FPA0P1_F),2))

###### Pruned

TNA0P0_P = cm_p[0][0]
FPA0P1_P = cm_p[0][1]
FNA1P0_P = cm_p[1][0]
TPA1P1_P = cm_p[1][1]


# Recall
recall_predict_noheartdisease_pruned = TNA0P0_P / (TNA0P0_P + FPA0P1_P) # divide by true =0
recall_predict_heartdisease_pruned = TPA1P1_P / (TPA1P1_P + FNA1P0_P) # divide by true =1

print('Pruned Tree Recall Predict No Heart Disease:', round(recall_predict_noheartdisease_pruned,2)) # predict no heart disease
print('Pruned Tree Recall Predict Heart Disease:', round(recall_predict_heartdisease_pruned,2)) #predict heart disease

# Accuracy
print('Pruned Tree Accuracy:', round(df_clf_prunedtree.score(x_test, y_test),2))

# Precision
print('Pruned Tree Precision:', round(TPA1P1_P/(TPA1P1_P + FPA0P1_P),2))
```

```python
# Decision Tree Bargraph on feature importance

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,8))
ax0 = fig.add_subplot(2,1,1) #2x2 grid, first area top left
ax1 = fig.add_subplot(2,1,2) #2x2 grid second area top right

ax0.set_xlim([0,0.5])
ax1.set_xlim([0,0.5])

ax0.barh(features, df_clf_fulltree.feature_importances_)
ax0.set_title('Feature Importance Full Tree')

ax1.barh(features, df_clf_prunedtree.feature_importances_)
ax1.set_title('Feature Importance Pruned Tree')

plt.show()
```

<b>Results:</b><br>
Our results (in the confusion matrix, and measures of accuracy, recall and precision) suggest little difference in reliability between a Full Tree (depth = 16) and a Pruned Tree (depth = 4).<br><br>
Based on statistical performance metrics from <b>Confusion Matrix</b>:<br>
Both trees have similar accuracies (70-72%) and precision (76-77%). <br>
Compared to the Full Tree, Pruned Tree had a lower recall predict for no Heart Disease (57% vs. 63%), but a higher recall predict for Heart Disease (81% vs. 74%).<br>

Based on <b>Feature Importance</b>:<br>
From the Full Tree, `chest_pain_type`, `cholesterol`, `maximum_heart_rate` are the top indicators in predicting heart disease.<br>

<b>c) <u>Logistic Regression</u></b><br>

Logistic Regression is a classification algorithm that provides a binary outcome with only two possible scenarios  either the event happens (1) or it does not happen (0). This is relevant for our investigation as the dependent variable (outcome) `diagnosis_of_heart_disease` is binary. It assumes no, or very little multicollinearity between the predictor variables (as verified in earlier Multivariate Analysis). The result allows us to calculate the various coefficents of each attribute, where we hope to find the best predictor of heart disease (highest coefficient). <br>

<b>Data Cleaning for Logistic Regression - Handling of Multi-Categorical Data:</b>
<br>
There are 3 data sets that were multi categorical: 
<br>
(i.e. data output consisted of a mix of values such as "0","1","2" representing different level of severities) <br>
- `chest_pain_type`<br>
- `resting_electrocardiographic_results`<br>
- `slope_peak_exercise`<br>

<br>
Multicategorical data would be acceptable for decision-tree analysis but not logistic regression. Thus, One-Hot Encoding was used to convert categorical data into numerical data to facilitate logistic regression.

```python
# one hot encode on categorical data, this is feed the machine learning algorithms to improve predictions
# column: Chest pain type
# column: Resting electrocardiographic results
# column: Slope peak exercise

CPT = pd.get_dummies(df['chest_pain_type'], prefix='CPT')
RER = pd.get_dummies(df['resting_electrocardiographic_results'], prefix='RER')
SPE = pd.get_dummies(df['slope_peak_exercise'], prefix='SPE')

df1 = pd.concat([df,CPT,RER,SPE], axis=1)


df1.drop(['chest_pain_type'],axis=1, inplace=True)
df1.drop(['resting_electrocardiographic_results'],axis=1, inplace=True)
df1.drop(['slope_peak_exercise'],axis=1, inplace=True)
```

```python
# train_test_split
import sklearn

x = df1[['age', 'sex', 'resting_blood_pressure',
       'cholesterol', 'fasting_blood_sugar', 'maximum_heart_rate',
       'exercise_induced_angina', 'ST_depression',
       'CPT_1.0', 'CPT_2.0', 'CPT_3.0',
       'CPT_4.0', 'RER_0.0', 'RER_1.0', 'RER_2.0', 'SPE_0.0', 'SPE_1.0',
       'SPE_2.0','SPE_3.0']]

y=df1[['diagnosis_of_heart_disease']]

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=5)
```

```python
#Methodology - logistic regression
# While loop choose which independent variables have internal corr > threshold=0.5

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


# LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)

print(y_test.shape)
print('Test accuracy:', round(accuracy_score(y_test, y_pred),2))

#The classifier has 78% cases correct, out of 184 test cases.
```

```python
# print out the various coefficents to identify which have the most positive relation to heart disease

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.barh(x_train.columns, logreg.coef_[0], color ='g')
plt.xlabel("Coefficient Values")
plt.ylabel("Risk Attributes")
plt.title("Coefficients for Logistic Regression")
plt.tight_layout() 
plt.show()

```

<h3>IV. Insights / Evaluation</h3><br>
<b>Insights from Overall Approach:</b><br>

Methods of replacing missing values may not be ideal.<br>

For <u>continuous data</u>, missing values were replaced with its <b>mean value</b> for the following:<br>

- `resting_blood_pressure`<br>
- `cholesterol`<br>
- `maximum_heart_rate`<br>
- `ST_depression`<br>

The missing data were replaced with the our sample mean. There was a possibility that it messes up the correlation of the attributes with having heart disease. The missing data could have been from healthy people as well as from people who have heart disease. Taking `cholesterol` for example, if a large proportion of people with missing data for `cholesterol` actually have heart disease, the expected `cholesterol` average for the missing data would be higher than the sample mean. This will skew the data for `cholesterol` and subsequently, the correlation of `cholesterol` and having heart disease.<br>

For <u>categorical data</u>, missing values were replaced with <b>"0"</b> for the following:<br>

- `fasting_blood_sugar`<br>
- `exercise_induced_angina`<br>
- `resting_electrocardiographic_results`<br>
- `slope_peak_exercise`

The issue is much worse for categorical data as it directly inserts data points that are either strictly healthy people or strictly with heart disease. For continuous data, when the replacement was done with the sample mean, it could somewhat capture the effects of both healthy and unhealthy people. Using `resting_electrocardiographic_results` as an example, having a value of 0 means that the `resting_electrocardiographic_results` are normal. People who actually have heart disease with a missing data point for the `resting_electrocardiographic_results` will have a replacement value of 0, indicating that the results are normal. However, this should not be the case. Clearly, it can be seen that the results will be skewed. This will decrease the reliability of these categorical attributes as a whole.

A way to circumnavigate this issue is to obtain more data points. By having more data points, the sample mean that is used to replace the missing data for continuous variables will be closer to the actual mean and more accurately reflect the attribute's correlation with having heart disease. Similarly for categorical data, the same issue will still be present. However, having more data points will dilute the effect of the inaccurate data, causing it to skew our results to a smaller extent. In addition, it unlocks the option for us to drop these data points entirely which is currently not possible due to the already limited dataset available.

With all attributes having weak correlations with each other and `diagnosis_of_heart_disease` (< 0.7), it also suggests that we should increase our sample size and experiment with a greater variety of more attributes (outside of the 14 original attributes) for further studies.

<h3>Glossary</h3>

<b>Fasting blood sugar</b>: 
A blood sample will be taken after an overnight fast. A fasting blood sugar level less than 100 mg/dL (5.6 mmol/L) is normal. A fasting blood sugar level from 100 to 125 mg/dL (5.6 to 6.9 mmol/L) is considered prediabetes. If it's 126 mg/dL (7 mmol/L) or higher on two separate tests, you have diabetes.

<b>METs Achieved</b>:
Metabolic Equivalent - A MET is a ratio of your working metabolic rate relative to your resting metabolic rate.

<b>Thal:</b>
Thalassemia (thal-uh-SEE-me-uh) is an inherited blood disorder that causes your body to have less haemoglobin than normal. Haemoglobin enables red blood cells to carry oxygen. Thalassemia can cause anemia, leaving you fatigued. If you have mild thalassemia, you might not need treatment.

<b>Exercise Induced Angina:</b>
Angina is a type of chest pain that results from reduced blood flow to the heart. A lack of blood flow means your heart muscle isn't getting enough oxygen. The pain is often triggered by physical activity or emotional stress. Stable angina, also called angina pectoris, is the most common type of angina. When one experiences chest pain that doesn't meet the criteria for angina, it's known as atypical chest pain. Non-angina pain describes chest pain that resembles heart pain in patients who do not have heart disease.

<b>Resting electrocardiographic results:</b>
The resting electrocardiogram is a test that measures the electrical activity of the heart. The heart is a muscular organ which pumps blood through rhythmic contractions induced by electric impulses generated by the sinus node, the hearts natural pacemaker. 

<b>ST depression:</b>
ST depression induced by exercise relative to rest. ST depression refers to a finding on an electrocardiogram, wherein the trace in the ST segment is abnormally low below the baseline.
